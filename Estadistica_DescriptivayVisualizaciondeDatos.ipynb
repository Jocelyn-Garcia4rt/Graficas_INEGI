```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

sizes = [201119, 195795, 58]
labels = ['Hombres', 'Mujeres', 'Sexo No Especificado']
colors = ['gold', 'lightblue', 'lightgreen']


mean = np.mean(sizes)
median = np.median(sizes)
std_dev = np.std(sizes)
mode_result = stats.mode(sizes)


if mode_result.count > 1:
    mode = mode_result.mode[0]
else:
    mode = "No hay moda"


print(f"Media: {mean}")
print(f"Mediana: {median}")
print(f"Desviación estándar: {std_dev}")
print(f"Moda: {mode}")


plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, wedgeprops={'edgecolor': 'black'})


plt.title('Diagrama de Pastel: Nacimientos de 1994 a 2023')

plt.figtext(0.15, 0.02, f'Media: {mean:.2f}', fontsize=12, color='black')
plt.figtext(0.15, 0.05, f'Mediana: {median:.2f}', fontsize=12, color='black')
plt.figtext(0.15, 0.08, f'Desviación Estándar: {std_dev:.2f}', fontsize=12, color='black')
plt.figtext(0.15, 0.11, f'Moda: {mode}', fontsize=12, color='black')

print("En base a nuestra grafica se puede ver que es minima la diferencia de los hombres y las mujeres nacidos, contrario a los de sexo no especificado. ")

plt.show()


```
