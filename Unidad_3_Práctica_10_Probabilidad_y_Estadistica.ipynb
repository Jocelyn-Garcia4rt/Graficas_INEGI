INTRODUCCIÓN

 La prueba de hipótesis es un metodo estadistico qque permite tomar decisiones sobre una población basándose en una muestra de datos. Este enfoque es esencial para validar afirmaciones o suposiciones sobre caracteristicas poblacionales, como medias o proporciones.

Parte 1: Calidad_Producto A y B

Calidad_Productos: Datos sobre dos lotes de productos, que se evaluarán para ver si
hay una diferencia significativa en la calidad (medida como calificación de calidad).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de los dos lotes
media_A = 73.9615  # Media del lote A
desviacion_A = 9.08168  # Desviación estándar del lote A
n_A = 100  # Tamaño del lote A

media_B = 78.2677  # Media del lote B (Ejemplo, debes colocar la real)
desviacion_B = 11.444  # Desviación estándar del lote B (Ejemplo, debes colocar la real)
n_B = 100  # Tamaño del lote B

alpha = 0.05  # Nivel de significancia

# Cálculo del estadístico t
numerador = media_A - media_B
denominador = np.sqrt((desviacion_A**2 / n_A) + (desviacion_B**2 / n_B))
t_calculado = numerador / denominador

# Cálculo de grados de libertad con la corrección de Welch
df = ((desviacion_A**2 / n_A) + (desviacion_B**2 / n_B))**2 / \
     (((desviacion_A**2 / n_A)**2 / (n_A - 1)) + ((desviacion_B**2 / n_B)**2 / (n_B - 1)))

# Valor crítico t para una prueba bilateral
t_critico = stats.t.ppf(1 - alpha / 2, df)

# Generación de la distribución t-Student
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Distribución t-Student", color='blue')

# Regiones de rechazo
x_rechazo_derecha = np.linspace(t_critico, 4, 100)
y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Región de rechazo")

x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

# Líneas de referencia
plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"t crítico = {t_critico:.2f}")
plt.axvline(x=-t_critico, color='black', linestyle='dashed')
plt.axvline(x=t_calculado, color='green', linestyle='dashed', label=f"t calculado = {t_calculado:.2f}")

# Etiquetas y leyenda
plt.title("Prueba t de dos muestras independientes")
plt.xlabel("Valor de t")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid()
plt.show()

# Resultados
print(f"Valor calculado de t: {t_calculado:.2f}")
print(f"Valor crítico de t (bilateral, α={alpha}): ±{t_critico:.3f}")
print(f"Grados de libertad ajustados: {df:.2f}")

# Decisión sobre la hipótesis
if abs(t_calculado) > t_critico:
    print("Rechazamos la hipótesis nula. Hay una diferencia significativa entre los dos productos.")
else:
    print("No rechazamos la hipótesis nula. No hay evidencia suficiente para afirmar que las ventas sean diferentes.")
```
