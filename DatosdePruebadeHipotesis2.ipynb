```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Cargar archivo Excel (subir manualmente en Colab antes de ejecutar)
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]  # Obtener nombre del archivo subido
sheets = pd.ExcelFile(file_name)
data = {sheet: sheets.parse(sheet) for sheet in sheets.sheet_names}

# Limpiar datos
calidad = data["Calidad_Productos"].iloc[:, :2].dropna()
ventas = data["Ventas"].iloc[:, 1:3].dropna()
tiempos = data["Tiempos_Entrega"].iloc[:, :2].dropna()
satisfaccion = data["Satisfaccion_Clientes"].dropna()

# Función para prueba T
def prueba_t(datos_a, datos_b, nombre):
    t_stat, p_valor = stats.ttest_ind(datos_a, datos_b, equal_var=False)
    decision = "Rechazamos H0" if p_valor < 0.05 else "No rechazamos H0"
    print(f"\n{nombre}")
    print(f"T-Statistic: {t_stat:.4f}, P-Valor: {p_valor:.4f}, Decisión: {decision}")
    graficar_t(t_stat, len(datos_a) + len(datos_b) - 2, nombre)
    return t_stat, p_valor

# Graficar una distribución t-Student para la prueba T
def graficar_t(t_stat, df, titulo="Distribución t-Student"):
    alpha = 0.05
    t_critico = stats.t.ppf(1 - alpha / 2, df)
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Distribución t-Student", color='blue')

    # Sombrar la región de rechazo
    x_rechazo_derecha = np.linspace(t_critico, 4, 100)
    y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
    plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Región de rechazo")

    x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
    y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
    plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

    # Líneas de referencia
    plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"T crítico = {t_critico:.2f}")
    plt.axvline(x=-t_critico, color='black', linestyle='dashed')
    plt.axvline(x=t_stat, color='green', linestyle='dashed', label=f"T calculado = {t_stat:.2f}")

    plt.title(titulo)
    plt.xlabel("Valor de T")
    plt.ylabel("Densidad de probabilidad")
    plt.legend()
    plt.grid()
    plt.show()

# Ejecutar pruebas T y graficar resultados
prueba_t(calidad["Lote_A"], calidad["Lote_B"], "Calidad de Productos")
prueba_t(ventas["Producto_A"], ventas["Producto_B"], "Ventas de Productos")
prueba_t(tiempos["Lote_A"], tiempos["Lote_B"], "Tiempos de Entrega")
```

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Cargar archivo Excel (subir manualmente en Colab antes de ejecutar)
from google.colab import files
uploaded = files.upload()

file_name = list(uploaded.keys())[0]  # Obtener nombre del archivo subido
sheets = pd.ExcelFile(file_name)
data = {sheet: sheets.parse(sheet) for sheet in sheets.sheet_names}

# Limpiar datos
calidad = data["Calidad_Productos"].iloc[:, :2].dropna()
ventas = data["Ventas"].iloc[:, 1:3].dropna()
tiempos = data["Tiempos_Entrega"].iloc[:, :2].dropna()
satisfaccion = data["Satisfaccion_Clientes"].dropna()

# Función para prueba T
def prueba_t(datos_a, datos_b, nombre):
    t_stat, p_valor = stats.ttest_ind(datos_a, datos_b, equal_var=False)
    decision = "Rechazamos H0" if p_valor < 0.05 else "No rechazamos H0"
    print(f"\n{nombre}")
    print(f"T-Statistic: {t_stat:.4f}, P-Valor: {p_valor:.4f}, Decisión: {decision}")
    graficar_t(t_stat, len(datos_a) + len(datos_b) - 2, nombre)
    return t_stat, p_valor, decision

# Graficar una distribución t-Student para la prueba T
def graficar_t(t_stat, df, titulo="Distribución t-Student"):
    alpha = 0.05
    t_critico = stats.t.ppf(1 - alpha / 2, df)
    x = np.linspace(-4, 4, 1000)
    y = stats.t.pdf(x, df)

    plt.figure(figsize=(8, 5))
    plt.plot(x, y, label="Distribución t-Student", color='blue')

    # Sombrar la región de rechazo
    x_rechazo_derecha = np.linspace(t_critico, 4, 100)
    y_rechazo_derecha = stats.t.pdf(x_rechazo_derecha, df)
    plt.fill_between(x_rechazo_derecha, y_rechazo_derecha, color='red', alpha=0.5, label="Región de rechazo")

    x_rechazo_izquierda = np.linspace(-4, -t_critico, 100)
    y_rechazo_izquierda = stats.t.pdf(x_rechazo_izquierda, df)
    plt.fill_between(x_rechazo_izquierda, y_rechazo_izquierda, color='red', alpha=0.5)

    # Líneas de referencia
    plt.axvline(x=t_critico, color='black', linestyle='dashed', label=f"T crítico = {t_critico:.2f}")
    plt.axvline(x=-t_critico, color='black', linestyle='dashed')
    plt.axvline(x=t_stat, color='green', linestyle='dashed', label=f"T calculado = {t_stat:.2f}")

    plt.title(titulo)
    plt.xlabel("Valor de T")
    plt.ylabel("Densidad de probabilidad")
    plt.legend()
    plt.grid()
    plt.show()

# Ejecutar pruebas T y almacenar resultados
resultados = []
resultados.append(("Calidad de Productos", *prueba_t(calidad["Lote_A"], calidad["Lote_B"], "Calidad de Productos")))
resultados.append(("Ventas de Productos", *prueba_t(ventas["Producto_A"], ventas["Producto_B"], "Ventas de Productos")))
resultados.append(("Tiempos de Entrega", *prueba_t(tiempos["Lote_A"], tiempos["Lote_B"], "Tiempos de Entrega")))

# Mostrar conclusiones individuales
print("\nCONCLUSIONES INDIVIDUALES:")
for nombre, t_stat, p_valor, decision in resultados:
    print(f"{nombre}: {decision}")

# Conclusión general
if any(decision == "Rechazamos H0" for _, _, _, decision in resultados):
    print("\nCONCLUSIÓN GENERAL: Se encontraron diferencias significativas en al menos una de las pruebas.")
else:
    print("\nCONCLUSIÓN GENERAL: No se encontraron diferencias significativas en ninguna de las pruebas.")
```
